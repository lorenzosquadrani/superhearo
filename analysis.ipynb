```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('audiogram_data.csv', header=[0, 1, 2], index_col=0)
data.index = pd.to_datetime(data.index, format='%d-%m-%Y' )

def string_to_list(string):
    return list(map(float, string.strip('[]').split(',')))

data = data.map(string_to_list)
data.head()
```

### All dates superposed

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 6))

axs[0, 0].set_title('Left ear, air conduction')
axs[0, 1].set_title('Right ear, air conduction')
axs[1, 0].set_title('Left ear, bone conduction')
axs[1, 1].set_title('Right ear, bone conduction')

for date in data.index:

    fs = data.loc[date, ('left', 'air', 'frequency')]
    dBs = data.loc[date, ('left', 'air', 'dB HL')]
    axs[0, 0].plot(fs, dBs,  'o--', label=date.strftime('%d-%m-%Y'))

    fs = data.loc[date, ('right', 'air', 'frequency')]
    dBs = data.loc[date, ('right', 'air', 'dB HL')]
    axs[0, 1].plot(fs, dBs, 'o--',  label=date.strftime('%d-%m-%Y'))

    fs = data.loc[date, ('left', 'bone', 'frequency')]
    dBs = data.loc[date, ('left', 'bone', 'dB HL')]
    axs[1, 0].plot(fs, dBs, 'o--',  label=date.strftime('%d-%m-%Y'))

    fs = data.loc[date, ('right', 'bone', 'frequency')]
    dBs = data.loc[date, ('right', 'bone', 'dB HL')]
    axs[1, 1].plot(fs, dBs, 'o--',  label=date.strftime('%d-%m-%Y'))    

for ax in axs.flatten():
    ax.set_xscale('log')
    ax.set_xlabel('Frequency (Hz)')
    ax.set_ylabel('dB HL')
    ax.yaxis.set_inverted(True)
    ax.set_ylim([60, 0])

fig.legend(axs[0,0].get_lines(), data.index.strftime('%d-%m-%Y'), loc='center right')
plt.tight_layout()
plt.show()
```

### Each frequency

```python
def get_data_for_frequency(frequency):

    dates_air, dates_bone = [], []
    dB_air_left, dB_air_right, dB_bone_left, dB_bone_right = [], [], [], []
    for date in data.index:

        frequencies_air = data.loc[date, ('left', 'air', 'frequency')]
        frequencies_bone = data.loc[date, ('left', 'bone', 'frequency')]

        if frequency in frequencies_air:
            dates_air.append(date)
            dB_air_left.append(data.loc[date, ('left', 'air', 'dB HL')][frequencies_air.index(frequency)])
            dB_air_right.append(data.loc[date, ('right', 'air', 'dB HL')][frequencies_air.index(frequency)])


        if frequency in frequencies_bone:
            dates_bone.append(date)
            dB_bone_left.append(data.loc[date, ('left', 'bone', 'dB HL')][frequencies_bone.index(frequency)])
            dB_bone_right.append(data.loc[date, ('right', 'bone', 'dB HL')][frequencies_bone.index(frequency)])
    
    return dates_air, dates_bone, dB_air_left, dB_air_right, dB_bone_left, dB_bone_right

def plot_frequency_data(frequency, ax):

    dates_air, dates_bone, dB_air_left, dB_air_right, dB_bone_left, dB_bone_right  = get_data_for_frequency(frequency)

    ax.set_title(f'Frequency: {frequency} Hz')
    ax.plot(dates_air, dB_air_left, 'o--', label='Left ear, air conduction', color='black')
    ax.plot(dates_air, dB_air_right, 'o--', label='Right ear, air conduction', color='red')
    ax.plot(dates_bone, dB_bone_left, 's--', label='Left ear, bone conduction', color='black')
    ax.plot(dates_bone, dB_bone_right, 's--', label='Right ear, bone conduction', color='red')

    ax.set_xlabel('Date')
    ax.set_ylabel('dB HL')
    ax.yaxis.set_inverted(True)
    ax.set_ylim([60, 0])
```

```python
frequencies = [125, 250, 500, 750, 1000, 1500, 2000, 3000, 4000, 6000, 8000]

fig, axs = plt.subplots(3, 4, figsize=(15, 10))

for frequency, ax in zip(frequencies, axs.flatten()[1:]):
    plot_frequency_data(frequency, ax)
    
    # set year format for x-axis
    ax.xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%m/%y'))

axs[0, 0].axis('off')
fig.legend(axs[0, 1].get_lines(), 
           ['Left ear, air conduction', 'Right ear, air conduction', 'Left ear, bone conduction', 'Right ear, bone conduction'],
            loc=(0.01, 0.8), fontsize=15)

plt.tight_layout()
plt.show()
```
