### Per frequency

```python
def get_data_for_frequency(frequency):

    """ 
    For a given frequency, loop over all the tests (dates), check whether that frequency was tested and if yes collect it.
    Return the collection of dates, air left measure, air right measure, bone left measure, bone right measure for the given frequency.
    """

    dates_air, dates_bone = [], []
    dB_air_left, dB_air_right, dB_bone_left, dB_bone_right = [], [], [], []

    for date in data.index:

        frequencies_air = data.loc[date, ('left', 'air', 'frequency')]
        frequencies_bone = data.loc[date, ('left', 'bone', 'frequency')]

        if frequency in frequencies_air:
            dates_air.append(date)
            dB_air_left.append(data.loc[date, ('left', 'air', 'dB HL')][frequencies_air.index(frequency)])
            dB_air_right.append(data.loc[date, ('right', 'air', 'dB HL')][frequencies_air.index(frequency)])


        if frequency in frequencies_bone:
            dates_bone.append(date)
            dB_bone_left.append(data.loc[date, ('left', 'bone', 'dB HL')][frequencies_bone.index(frequency)])
            dB_bone_right.append(data.loc[date, ('right', 'bone', 'dB HL')][frequencies_bone.index(frequency)])
    
    return dates_air, dates_bone, dB_air_left, dB_air_right, dB_bone_left, dB_bone_right

def plot_frequency_data(frequency, ax_l, ax_r):

    dates_air, dates_bone, dB_air_left, dB_air_right, dB_bone_left, dB_bone_right  = get_data_for_frequency(frequency)

    ax.set_title(f'Frequency: {frequency} Hz')

    ax_l.plot(dates_air, dB_air_left, 'o--', label='Left ear, air conduction', color='tab:blue')
    ax_l.plot(dates_bone, dB_bone_left, 's--', label='Left ear, bone conduction', color='black')

    c = 'red'
    ax_r.plot(dates_air, dB_air_right, 'o--', label='Right ear, air conduction', color='tab:blue')
    ax_r.plot(dates_bone, dB_bone_right, 's--', label='Right ear, bone conduction', color='black')

    ax_l.axvline(pd.Timestamp('2024-06-15'), color='red', linestyle='--')
    ax_r.axvline(pd.Timestamp('2024-06-15'), color='red', linestyle='--')

    ax_l.yaxis.set_inverted(True)
    ax_r.yaxis.set_inverted(True)
    ax_r.set_xticks([pd.Timestamp('2021-01-01'),pd.Timestamp('2023-01-01'), pd.Timestamp('2025-01-01')])
    ax_r.xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%Y'))
```

```python
frequencies = [125, 250, 500, 750, 1000, 1500, 2000, 3000, 4000, 6000, 8000]

fig, axs = plt.subplots(2, len(frequencies), figsize=(18,3), sharex=True, sharey=True, dpi=300)


for frequency, (ax_l, ax_r) in zip(frequencies, axs.T):

    ax_l.set_title('{}kHz'.format(frequency/1000))

    plot_frequency_data(frequency, ax_l, ax_r)

axs[0, 0].set_ylabel('Left')
axs[1,0].set_ylabel('Right')

plt.savefig('per_frequency.png', dpi=300)
plt.show()
```
